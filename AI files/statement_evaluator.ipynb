```python
import pandas as pd
import numpy as np


dataFrameCSV = pd.read_csv("archive/userDataTable.csv")
dataFrameCSV.head()
```

```python
dataFrameCSV.iloc[: , -1] = dataFrameCSV.iloc[: , -1].str.replace(',','')
lastColumn = dataFrameCSV.iloc[: , -1].dropna()

Balance = lastColumn.astype(float)
Balance
```

```python
print(len(Balance))
```

```python
loan_amount = 1000
  
count = 0
for i in Balance :
    if i < loan_amount :
        count = count + 1

print ("The numbers lesser than loan amount:" + str(loan_amount)+ "N is " + str(count)+ " out of " + str(len(Balance)))
```

```python
import math
balance_length = len(Balance)

quotient = (balance_length/3)
valid_percent = math.floor(quotient)
print(valid_percent)
```

```python
if count > valid_percent:
    print("Don't give")
else:
    print("Give")
```
